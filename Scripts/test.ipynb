```python
import psyneulink as pnl
input_ = pnl.TransferMechanism(input_shapes=2)
map_ = pnl.MappingProjection(matrix=pnl.IDENTITY_MATRIX)
output_ = pnl.TransferMechanism(input_shapes=2)

comp_ = pnl.AutodiffComposition(
    pathways=[(input_, map_, output_)]
)

comp_.show_graph(output_fmt='jupyter')
```
