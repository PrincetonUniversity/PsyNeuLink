# Mechanism - Demo

Mechanisms take inputs, transform them (often via a function) and make the result available as output. Here, we will explore the basic functionality of mechanisms in PsyNeuLink by using the most used mechanism, the `ProcessingMechanism`.

```python
import psyneulink as pnl
```

## Creating a ProcessingMechanism

To create a mechanism, we call its constructor:


```python
example_mech = pnl.ProcessingMechanism()
```

We can "run" this mechanism on input using the execute function or add it to a composition and run the composition:

```python
example_mech.execute(1)
```

```python
comp = pnl.Composition()
comp.add_node(example_mech)

comp.run([1])
```

We see, that the "default" transformation (or `function`) for the mechanism is the identity function, which simply returns the input as output. We can inspect the function of a mechanism by accessing the `function` attribute:

```python
example_mech.function
```
